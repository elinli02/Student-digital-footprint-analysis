```python
import csv
import pandas as pd
data=pd.read_csv('/content/_select_distinct_sv_students_id_as_students_id_g_id_as_group_id__202202171014.csv', delimiter=',')
data
```

```python
#подсчет значений посещаемости
data['visit'] = pd.to_numeric(data['visit'], errors='coerce')
data['visit'].value_counts()
```

```python
data.groupby(['students_id']).agg({'lesson_types_id': "count", 'visit': "sum"})
```

```python
data.groupby(['students_id'])['visit'].agg('sum').count()
```

```python
#количество посещений лекций
lecture_1 = data.query('lesson_types_id == 28 & visit == 1')
lecture_1.count()
```

```python
#количество пропуска лекций
lecture_0 = data.query('lesson_types_id == 28 & visit == 0')
lecture_0.count()
```

```python
#количество посещений практик
practical_1 = data.query('lesson_types_id == 35 & visit == 1')
practical_1.count()
```

```python
#1 Зависимость "id дисциплины" + посещаемость
#посещение конкретных дисциплин
dis = data.groupby(['discipline_id']).agg({'visit': "sum"})
#dis
dis.sort_values(by=['visit'], ascending=[False])
```

```python
#2 Зависимость "Тип пары" + посещаемость
type_visit = data.groupby(['lesson_types_name']).agg({'visit': "sum"})
type_visit
```

```python
#3 Зависимость "Номер пары" + посещаемость
num_less = data.groupby(['lesson_num']).agg({'visit': "sum"})
num_less
```

```python
#подсчет посещений по датам
date_visit = data.groupby(['lessons_date']).agg({'visit': "sum"})
date_visit
```

```python
#3 (в прошлой работе зависимость "Номер пары + посещаемость")

#4 зависимость "Дата пары(месяц)" + посещаемость
import datetime 

data['lessons_date'] = pd.to_datetime(data['lessons_date'], errors='coerce')
data['month'] = data['lessons_date'].dt.month
data.groupby(['month']).agg({'visit': "sum"})
```
